```python
import os 
import numpy as np 
import cv2
from matplotlib import pyplot as plt

faceClasssifier = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
data_path = 'dataset/train_data'
categories = os.listdir(data_path)
```

```python
categories
```

```python
labels = [i for i in range(len(categories))]
```

```python
category_dict = dict(zip(categories, labels))
category_dict
```

```python
data = []
target = []

for category in categories:
    upper_path = os.path.join(data_path, category)
    image_names = os.listdir(upper_path)
    for image_name in image_names:
        image_path = os.path.join(upper_path,image_name)
        img = cv2.imread(image_path)
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        faces = faceClasssifier.detectMultiScale(gray) 
        for (x, y, w, h) in faces:
            cropped_face = img[y:y+h, x:x+w]
            cv2.imshow('cropped_face', cropped_face)
            print("accept - y reject - n")
            k = cv2.waitKey(0)
            if k == 121:
                cropped_face_gray = cv2.cvtColor(cropped_face, cv2.COLOR_BGR2GRAY)
                cropped_face_gray = cv2.resize(cropped_face_gray, (50, 50))
                data.append(cropped_face_gray)
                target.append(category_dict[category])
cv2.destroyAllWindows()
        
```

```python
data = np.array(data)
target = np.array(target)
```
